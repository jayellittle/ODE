# Solving ordinary differential equation by numerical analysis

## Euler method

### Free Gravitation Fall

Define function `euler(x0, v0)`, which return `x1` and `v1`.

- `x1` : Value of $x$ after time $dt$ passed

$$
x_{i+1} = x_i + \frac{dx}{dt} dt
$$

- `v1` : Value of $v$ after time $dt$ passed

$$
\begin{align*}
v_{i+1} & = v_i + \frac{dv}{dt} dt\\
        & = v_i - g~dt
\end{align*}
$$

```python
def euler(x0, v0):
    v1 = v0 - g * dt
    x1 = x0 + v0 * dt
    return x1, v1
```

Define function `my_plot(xx, vv, tt)`, which plot diagrams of height(`xx`) and velocity(`vv`)

- `xx` : List of height values per time

- `vv` : List of velocity values per time

- `tt` : List of time values per $dt$

```python
import matplotlib.pyplot as plt

def my_plot(xx, vv, tt):
    plt.plot(tt, xx, color='b', linestyle='--', label="height")
    plt.plot(tt, vv, color='r', label="velocity")
    plt.legend()
    plt.xlabel("time")
    plt.ylabel("height and velocity")
    plt.grid()
    plt.show()
```

Set initial values of each variables

- `g` : Gravitation acceleration $g$

- Value inside list `tt`, `xx`, `vv` are the initial values

  Upon the following code, start on time $t=0$, on location $(x,y)=(10,0)$

```python
g, dt = 9.8, 0.01
tt, xx, vv = [0.0], [10.0], [0.0]
t = 0.0
```

```python
for i in range(0, 250):
    t += dt
    x, v = euler(xx[-1], vv[-1])
    tt.append(t)
    xx.append(x)
    vv.append(v)

my_plot(xx, vv, tt)
```

### Applying air resistance

Define `euler2(x0, v0)` which return `x1` and `v1` with air resistance applied

$$
v_{i+1} = v_i - (Cv_i + g)dt
$$

- `cc` : Drag coefficient

```python
def euler2(x0, v0):
    v1 = v0 + (-cc * v0 - g) * dt
    x1 = x0 + v0 * dt
    return [x1, v1]
```

```python
g, dt, cc = 9.8, 0.01, 1.0
tt, xx, vv = [0.0], [10.0], [0.0]
t = 0.0
```

```python
for i in range(0, 500):
    t += dt
    x, v = euler2(xx[-1], vv[-1])
    tt.append(t)
    xx.append(x)
    vv.append(v)

my_plot(xx, vv, tt)
```

### Free oscillation mass-spring system

Motion of mass-spring system could be defined as following function `euler3(x0, v0)`, as same as other systems.

```python
def euler3(x0, v0):
    v1 = v0 + (-k * x0) * dt
    x1 = x0 + v0 * dt
    return [x1, v1]
```

After declaring initial values as following, the diagram following next is what we get.

```python
t, dt, k = 0.0, 0.01, 0.01
tt, xx, vv = [0.0], [0.0], [0.01]
for i in range(0, 100000):
    t += dt
    x, v = euler3(xx[-1], vv[-1])
    tt.append(t)
    xx.append(x)
    vv.append(v)

my_plot(xx, vv, tt)
```

By looking carefully on values of height, it's growing bigger than the maximum value it's supposed to get. This is because, Euler method is relatively a 'simple' method of numerical analyzing, and its error is bigger than what we think.
As an alternative, **Runge-Kutta method** is recommended to use.
